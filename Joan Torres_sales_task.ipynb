```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
month=np.array(["Jan","Feb","Mar"])
pa=np.array([120,130,125])
pb=np.array([150,160,170])
pc=np.array([100,110,105])
msales=pd.DataFrame(data=[pa,pb,pc],columns=["Product A","Product B","Product C"],index=month)
print(msales)
msales.describe()
```

# explanation

mean() it is used to calculare the average in this case the average for the product sales per month.

min () it is the minimum value showed in the data frame, for this data frame it is the minimum sale.

count() it return how many times a data apears in the data frame, for this scenario each product appears 3 times per month.

```python
plt.plot(month,pa,label="Product A")
plt.plot(month,pb,label="Product B")
plt.plot(month,pc,label="Product C")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.title("Product Sales")
plt.legend()
plt.show()
```

# Interpretion the data for sales A
Products A, B and C started with great acceptance in the market during the beginning of the year, but by February we noticed that products B and C kept descending sales.
